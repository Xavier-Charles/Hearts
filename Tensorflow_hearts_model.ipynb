```python
# select and manipulate features
def preprocess_features(hearts_data):
    selected_features =  hearts_data[['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol', 'Fbs', 'RestECG', 'MaxHR',
       'ExAng', 'Oldpeak', 'Slope', 'Ca', 'Thal']]
    processed_features = selected_features.copy()  
    # Create a synthetic feature. 
    return processed_features
```

```python
# select and manipulate targets
def preprocess_targets(hearts_data):  
    output_targets = hearts_data[['AHD']]
    return output_targets
```

```python
# randomizing the data
hearts_data = hearts_data.reindex(    
    np.random.permutation(hearts_data.index))
```

```python
# Choose the first 230 (out of 303) examples for training. 
training_examples = preprocess_features(hearts_data.head(230)) 
training_targets = preprocess_targets(hearts_data.head(230)) 
test_examples = preprocess_features(hearts_data.tail(73)) 
test_targets = preprocess_targets(hearts_data.tail(73))
```

```python
# Check that we've done the right thing. 
print("Training examples summary:",display.display(training_examples.describe())) 
print("Training targets summary:",display.display(training_targets.describe()))
```

```python
# defining functions

def construct_feature_columns(input_features):
  """Args:
    input_features: The names of the numerical input features to use.
  Returns:
    A set of feature columns
  """ 
  return set([tf.feature_column.numeric_column(my_feature)
              for my_feature in input_features])



def input_fn(features, targets, num_epochs=None, shuffle=True):
  
   # Convert pandas data into a dict of np arrays.
    features = {key:np.array(value) for key,value in dict(features).items()}                                             
 
    # Construct a dataset, and configure batching/repeating.
    ds = Dataset.from_tensor_slices((features,targets)) # warning: 2GB limit
    ds = ds.batch(10).repeat(num_epochs)  #batch_size
    
    # Shuffle the data, if specified.
    if shuffle:
      ds = ds.shuffle(10000)
    
    # Return the next batch of data.
    features, labels = ds.make_one_shot_iterator().get_next()
    return features, labels
```
